```python
print("\n--- Analyse Univariée : Visualisation des Distributions ---")

# Variables Numériques : Histograms et Box Plots
numeric_cols = ['TransactionAmount', 'CustomerAge', 'TransactionDuration', 'LoginAttempts', 'AccountBalance']
for col in numeric_cols:
    print(f"\nDistribution de {col}:")
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1) # Histogramme
    sns.histplot(df[col].dropna(), kde=True) # kde=True ajoute une courbe de densité
    plt.title(f'Histogramme de {col}')

    plt.subplot(1, 2, 2) # Box Plot
    sns.boxplot(x=df[col].dropna())
    plt.title(f'Box Plot de {col}')

    plt.tight_layout()
    plt.show()

# Variables Catégorielles : Bar Plots ou Value Counts
categorical_cols_low_card = ['TransactionType', 'Channel', 'CustomerOccupation']
for col in categorical_cols_low_card:
    print(f"\nDistribution de {col}:")
    plt.figure(figsize=(8, 5))
    sns.countplot(y=df[col].dropna(), order=df[col].value_counts().index) # Ordonner par fréquence
    plt.title(f'Fréquence de {col}')
    plt.show()

# Variables Catégorielles : Top N pour haute cardinalité
high_card_cols = ['Location', 'DeviceID', 'IP Address', 'MerchantID']
for col in high_card_cols:
     print(f"\nTop 10 des valeurs dans {col}:")
     print(df[col].value_counts().head(10))
```

```python
 #pour chaque client (AccountID), on doit mettre toutes ses transactions dans l'ordre du temps (TransactionDate). C'est crucial pour calculer le "temps écoulé depuis la précédente transaction"
 #On crée une nouvelle variable df_sorted pour ne pas mélanger l'ordre de votre DataFrame original si vous en aviez besoin dans l'ordre d'origine plus tard dans le même notebook d'EDA. 
 df_sorted = df.sort_values(by=['AccountID', 'TransactionDate'])
```

```python
    # Calculer la différence de temps entre chaque transaction et la précédente POUR CHAQUE COMPTE
    # On "groupe" d'abord par AccountID pour que le diff() ne compare pas une transaction du compte A
    # avec la dernière transaction du compte B.
    df_sorted['TimeDelta'] = df_sorted.groupby('AccountID')['TransactionDate'].diff()
```

```python
    # Convertir cette différence de temps en nombre total de secondes
    df_sorted['TimeSinceLastTransaction_s'] = df_sorted['TimeDelta'].dt.total_seconds()
```

```python
    # --- Dans 1_EDA.ipynb, après avoir créé la colonne ---
    print("\nStatistiques descriptives pour le temps écoulé depuis la dernière transaction (secondes) :")
    print(df_sorted['TimeSinceLastTransaction_s'].describe())

    print("\nDistribution du temps écoulé (secondes) :")
    # Limiter l'affichage pour voir les intervalles courts si besoin
    plt.figure(figsize=(10, 6))
    sns.histplot(df_sorted['TimeSinceLastTransaction_s'].dropna(), bins=50) # ignore les valeurs vides
    # plt.xlim(0, 3600*24) # Optionnel : Limiter l'axe X pour mieux voir les durées courtes (ici, transactions de moins d'un jour)
    plt.title('Distribution du Temps écoulé depuis la dernière transaction (s)')
    plt.xlabel('Secondes')
    plt.ylabel('Nombre de transactions')
    plt.show()
```
